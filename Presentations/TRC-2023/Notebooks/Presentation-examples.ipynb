```raku
sub solve_quadratic_equation($a, $b, $c) {
    my $discriminant = $b ** 2 - 4 * $a * $c;

    if $discriminant > 0 {
        my $x1 = (-$b + sqrt($discriminant)) / (2 * $a);
        my $x2 = (-$b - sqrt($discriminant)) / (2 * $a);
        return ($x1, $x2);
    }
    elsif $discriminant == 0 {
        my $x = -$b / (2 * $a);
        return $x;
    }
    else {
        return ();  # No real solutions
    }
}
```
